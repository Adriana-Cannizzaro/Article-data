```python
# regression plot using seaborn
fig = plt.figure(figsize=(10,7))
                        
sns.regplot(x=df_AD.NPIDSEV, y=df_AD.temporal_entorhinal_ThickL_DK, color='blue')
sns.regplot(x=df_MCI.NPIDSEV, y=df_MCI.temporal_entorhinal_ThickL_DK, color='green')

# legend, title, and labels.
plt.legend(labels=['AD', 'MCI'])
plt.title('Impact of depression on entorhinal thickness in AD and MCI', size=24)
plt.xlabel('NPIDSEV (depression)', size=18)
plt.ylabel('temporal_entorhinal_ThickL_DK', size=18);
```
